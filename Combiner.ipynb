```python
import pandas as pd
```

```python
def loadCSV(filename):
    return pd.read_csv(filename)
```

```python
def combine(prefix):
    s_1=loadCSV(prefix+"_0_100_simulation.csv")
    s_2=loadCSV(prefix+"_100_200_simulation.csv")
    s_3=loadCSV(prefix+"_200_300_simulation.csv")
    s_4=loadCSV(prefix+"_300_400_simulation.csv")
    s_5=loadCSV(prefix+"_400_500_simulation.csv")
    return pd.concat([s_1, s_2, s_3, s_4, s_5])
```
